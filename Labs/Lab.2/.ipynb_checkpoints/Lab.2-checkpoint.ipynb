*Exercise 2:* Write a function that takes two integers `n` and `m` and **draws** an `n` by `m` game board.

For example, the following is a 3×3 board:

```
   --- --- --- 
  |   |   |   | 
   --- --- ---  
  |   |   |   | 
   --- --- ---  
  |   |   |   | 
   --- --- --- 
```


````python
def draw_empty(n, m):
    top = "   " + ("--- " * m)
    mid = "  |" + ("   |" * m)
    for _ in range(n):
        print(top)
        print(mid)
    print(top)
````

*Exercise 3:* Modify Exercise 2 so that it takes a matrix in the format from Exercise 1 and draws a tic‑tac‑toe board with `"X"`s and `"O"`s.

````python
def draw_from_matrix(b):
    n = len(b)
    m = len(b[0]) if n else 0
    sym = {0: " ", 1: "X", 2: "O"}

    top = " " + ("--- " * m)
    for r in range(n):
        print(top)
        print(" |" + (" |".join(f" {sym[b[r][c]]} " for c in range(m))) + " |")
    print(top)
````

*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
def game_status(b):
    n = len(b)

    def all_same(line, p):
        return all(x == p for x in line)

    for p in (1, 2):
        for r in range(n):
            if b[r][0] != 0 and all_same(b[r], p):
                return p
        for c in range(n):
            col = [b[r][c] for r in range(n)]
            if col[0] != 0 and all_same(col, p):
                return p

        d1 = [b[i][i] for i in range(n)]
        d2 = [b[i][n-1-i] for i in range(n)]
        if d1[0] != 0 and all_same(d1, p):
            return p
        if d2[0] != 0 and all_same(d2, p):
            return p

    return -1 if any(0 in row for row in b) else 0
````

*Exercise 5:* Write a function that takes a game board, a player number, and `(row, col)` coordinates and places the correct mark (`"X"` or `"O"`) in that location.

Requirements:

- Only allow placing a mark in a previously empty location.
- Return `True` if the move was successful, and `False` otherwise.


````python
def place_mark(board, player, row, col):
    if board[row][col] != 0:
        return False
    board[row][col] = player
    return True
````

````python
b = make_board(3)

print(place_mark(b, 1, 0, 0)) 
print(place_mark(b, 2, 0, 0)) 
print(place_mark(b, 2, 1, 1)) 
print(b)
````

*Exercise 6:* Modify Exercise 3 to show **row and column labels** so that players can specify locations like `"A2"` or `"C1"`.

````python
def draw_labeled(board):
    n = len(board)
    m = len(board[0]) if n else 0
    sym = {0: " ", 1: "X", 2: "O"}

    # column labels
    print("   " + " ".join(f" {chr(65+c)} " for c in range(m)))

    top = " " + ("--- " * m)
    for r in range(n):
        print(top)
        print(f"{r+1:>2}|" + "|".join(f" {sym[board[r][c]]} " for c in range(m)) + "|")
    print(top)
````

*Exercise 7:* Write a function that takes a board, a player number, and a location string (as in Exercise 6), then uses your function from Exercise 5 to update the board.

````python
def apply_location(board, player, loc):
    loc = loc.strip().upper()

    if len(loc) < 2:
        return False

    col = ord(loc[0]) - ord('A')
    row = int(loc[1:]) - 1

    if row < 0 or col < 0 or row >= len(board) or col >= len(board):
        return False

    return place_mark(board, player, row, col)
````

````python
b = make_board(3)

print(apply_location(b, 1, "A1"))  
print(apply_location(b, 2, "A1"))  
print(apply_location(b, 2, "C3"))  
draw_labeled(b)
````

*Exercise 8:* Write a function that is called with a board and player number, takes input from the player using Python's `input()`, and modifies the board using your function from Exercise 7.

Keep asking for input until the player enters a valid location that results in a valid move.


````python
def player_turn(board, player):
    while True:
        loc = input(f"Player {player}, enter location (e.g. A2): ")
        if apply_location(board, player, loc):
            return
        print("Invalid move. Please Try again.")
````

````python
b = make_board(3)
draw_labeled(b)
player_turn(b, 1)
draw_labeled(b)
````

*Exercise 9:* Use all of the previous exercises to implement a full tic‑tac‑toe game:

- draw the board,
- repeatedly ask two players for a location,
- apply valid moves,
- check the game status until a player wins or the game is a draw.


````python
def play_game(n):
    board = make_board(n)
    player = 1

    while True:
        draw_labeled(board)

        s = game_status(board)
        if s == 1:
            print("Player 1 (X) wins!")
            return
        if s == 2:
            print("Player 2 (O) wins!")
            return
        if s == 0:
            print("Draw.")
            return

        player_turn(board, player)
        player = 2 if player == 1 else 1
````

*Exercise 11:* Develop a version of the game where one player is the computer.

Note: you do **not** need an extensive search for the best move. For example, you can have the computer:
- block obvious losses
- otherwise try to create a winning row/column/diagonal


````python
import random

def computer_move(board, player):
    n = len(board)
    opp = 2 if player == 1 else 1

    empty = [(r, c) for r in range(n) for c in range(n) if board[r][c] == 0]

    # try winning move
    for r, c in empty:
        test = [row[:] for row in board]
        test[r][c] = player
        if game_status(test) == player:
            place_mark(board, player, r, c)
            return

    # try blocking move
    for r, c in empty:
        test = [row[:] for row in board]
        test[r][c] = opp
        if game_status(test) == opp:
            place_mark(board, player, r, c)
            return

    # otherwise: center/corners bias, but random within choices
    center = [(n//2, n//2)] if n % 2 == 1 else []
    corners = [(0,0), (0,n-1), (n-1,0), (n-1,n-1)]

    choices = [m for m in center if m in empty] or [m for m in corners if m in empty] or empty
    r, c = random.choice(choices)
    place_mark(board, player, r, c)
````

````python
def play_vs_computer(n):
    board = make_board(n)
    player = 1  # player 1 = human, player 2 = computer

    while True:
        draw_labeled(board)

        s = game_status(board)
        if s == 1:
            print("Player 1 (X) wins!")
            return
        if s == 2:
            print("Computer (O) wins!")
            return
        if s == 0:
            print("Draw.")
            return

        if player == 1:
            player_turn(board, 1)
        else:
            computer_move(board, 2)

        player = 2 if player == 1 else 1

play_vs_computer(3)
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
import copy, random

def score_board(board, player_to_win):
    s = game_status(board)
    if s == player_to_win:
        return 10
    if s == 0:
        return 0
    if s in (1, 2):
        return -10
    return 0

def generate_moves(board, player_to_win, current_player, depth):
    if depth == 0 or game_status(board) != -1:
        return score_board(board, player_to_win)

    scores = []
    n = len(board)

    for r in range(n):
        for c in range(n):
            if board[r][c] == 0:
                new_board = copy.deepcopy(board)
                place_mark(new_board, current_player, r, c)

                sc = generate_moves(
                    new_board,
                    player_to_win,
                    2 if current_player == 1 else 1,
                    depth - 1
                )
                scores.append(sc)

    if current_player == player_to_win:
        return max(scores)
    else:
        return min(scores)

def computer_move_search(board, player, depth=5):
    n = len(board)
    best_score = -float("inf")
    best_moves = []

    for r in range(n):
        for c in range(n):
            if board[r][c] == 0:
                new_board = copy.deepcopy(board)
                place_mark(new_board, player, r, c)

                sc = generate_moves(
                    new_board,
                    player,
                    2 if player == 1 else 1,
                    depth - 1
                )

                if sc > best_score:
                    best_score = sc
                    best_moves = [(r, c)]
                elif sc == best_score:
                    best_moves.append((r, c))

    r, c = random.choice(best_moves)
    place_mark(board, player, r, c)
````

````python
def play_vs_computer_search(n, depth=5):
    board = make_board(n)
    player = 1

    while True:
        draw_labeled(board)

        s = game_status(board)
        if s == 1:
            print("Player 1 (X) wins!")
            return
        if s == 2:
            print("CPU (O) wins!")
            return
        if s == 0:
            print("Draw.")
            return

        if player == 1:
            player_turn(board, 1)
        else:
            computer_move_search(board, 2, depth)

        player = 2 if player == 1 else 1

play_vs_computer_search(3, depth=5)
````
