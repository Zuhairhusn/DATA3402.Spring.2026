1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value):
        self.value = 0
        self.max_value = max_value

    def increment(self):
        if self.value + 1 > self.max_value:
            print("error")
        else:
            self.value += 1

    def reset(self):
        self.value = 0


# test
c = Counter(3)
print(c.value)
c.increment(); c.increment(); c.increment()
print(c.value)
c.increment()  # error
c.reset()
print(c.value)
```

2. Copy and paste your solution to question 1 and modify it so that all the data held by the counter is private. Implement functions to check the value of the counter, check the maximum value, and check if the counter is at the maximum.

```python
class Counter:
    def __init__(self, max_value):
        self.__value = 0
        self.__max_value = max_value

    def increment(self):
        if self.__value + 1 > self.__max_value:
            print("error")
        else:
            self.__value += 1

    def reset(self):
        self.__value = 0

    def get_value(self):
        return self.__value

    def get_max(self):
        return self.__max_value

    def at_max(self):
        return self.__value == self.__max_value


# test
c = Counter(2)
c.increment()
print(c.get_value())
print(c.get_max())
print(c.at_max())
c.increment()
print(c.at_max())
c.increment()  # error
```

6. Implement a triangle class analogous to the rectangle and circle in question 5.

```python
import math

class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3):
        self.__p1 = (x1, y1)
        self.__p2 = (x2, y2)
        self.__p3 = (x3, y3)

    def area(self):
        x1,y1 = self.__p1
        x2,y2 = self.__p2
        x3,y3 = self.__p3
        return abs((x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)) / 2)

    def perimeter(self):
        def d(a,b): return math.hypot(a[0]-b[0], a[1]-b[1])
        return d(self.__p1,self.__p2) + d(self.__p2,self.__p3) + d(self.__p3,self.__p1)

    def get_p1(self): return self.__p1
    def get_p2(self): return self.__p2
    def get_p3(self): return self.__p3


# test
t = Triangle(0,0, 4,0, 0,3)
print(t.get_p1(), t.get_p2(), t.get_p3())
print(t.area())
print(t.perimeter())
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object. 

```python
import math

class Shape:
    def area(self): raise NotImplementedError
    def perimeter(self): raise NotImplementedError
    def points(self, n=16): raise NotImplementedError


class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def area(self): return self.__length * self.__width
    def perimeter(self): return 2 * (self.__length + self.__width)

    def points(self, n=16):
        n = max(4, min(16, int(n)))
        x0, y0 = self.__x, self.__y
        x1, y1 = self.__x + self.__length, self.__y + self.__width
        return [(x0,y0),(x1,y0),(x1,y1),(x0,y1)][:n]

    def get_length(self): return self.__length
    def get_width(self): return self.__width
    def get_x(self): return self.__x
    def get_y(self): return self.__y


class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def area(self): return math.pi * self.__radius ** 2
    def perimeter(self): return 2 * math.pi * self.__radius

    def points(self, n=16):
        n = max(4, min(16, int(n)))
        pts = []
        for i in range(n):
            a = 2 * math.pi * i / n
            pts.append((self.__x + self.__radius*math.cos(a),
                        self.__y + self.__radius*math.sin(a)))
        return pts

    def get_radius(self): return self.__radius
    def get_x(self): return self.__x
    def get_y(self): return self.__y


class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3):
        self.__p1=(x1,y1); self.__p2=(x2,y2); self.__p3=(x3,y3)

    def area(self):
        x1,y1=self.__p1; x2,y2=self.__p2; x3,y3=self.__p3
        return abs((x1*(y2-y3)+x2*(y3-y1)+x3*(y1-y2))/2)

    def perimeter(self):
        def d(a,b): return math.hypot(a[0]-b[0], a[1]-b[1])
        return d(self.__p1,self.__p2)+d(self.__p2,self.__p3)+d(self.__p3,self.__p1)

    def points(self, n=16):
        n = max(3, min(16, int(n)))
        return [self.__p1, self.__p2, self.__p3][:n]


# test
r = Rectangle(5,3,0,0)
c = Circle(3,0,0)
t = Triangle(0,0,4,0,0,3)
print(r.points())
print(len(c.points()), c.points()[:3])
print(t.points())
```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
import math

class Shape:
    def area(self): raise NotImplementedError
    def perimeter(self): raise NotImplementedError
    def points(self, n=16): raise NotImplementedError
    def contains(self, x, y): raise NotImplementedError


class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length=length; self.__width=width; self.__x=x; self.__y=y

    def area(self): return self.__length*self.__width
    def perimeter(self): return 2*(self.__length+self.__width)

    def points(self, n=16):
        x0,y0=self.__x,self.__y
        x1,y1=self.__x+self.__length, self.__y+self.__width
        return [(x0,y0),(x1,y0),(x1,y1),(x0,y1)][:max(4,min(16,int(n)))]

    def contains(self, x, y):
        return self.__x <= x <= self.__x+self.__length and self.__y <= y <= self.__y+self.__width


class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__r=radius; self.__x=x; self.__y=y

    def area(self): return math.pi*self.__r**2
    def perimeter(self): return 2*math.pi*self.__r

    def points(self, n=16):
        n=max(4,min(16,int(n)))
        return [(self.__x+self.__r*math.cos(2*math.pi*i/n),
                 self.__y+self.__r*math.sin(2*math.pi*i/n)) for i in range(n)]

    def contains(self, x, y):
        return (x-self.__x)**2 + (y-self.__y)**2 <= self.__r**2


class Triangle(Shape):
    def __init__(self,x1,y1,x2,y2,x3,y3):
        self.__p1=(x1,y1); self.__p2=(x2,y2); self.__p3=(x3,y3)

    def area(self):
        x1,y1=self.__p1; x2,y2=self.__p2; x3,y3=self.__p3
        return abs((x1*(y2-y3)+x2*(y3-y1)+x3*(y1-y2))/2)

    def perimeter(self):
        def d(a,b): return math.hypot(a[0]-b[0],a[1]-b[1])
        return d(self.__p1,self.__p2)+d(self.__p2,self.__p3)+d(self.__p3,self.__p1)

    def points(self, n=16):
        n=max(3,min(16,int(n)))
        return [self.__p1,self.__p2,self.__p3][:n]

    def contains(self, x, y):
        x1,y1=self.__p1; x2,y2=self.__p2; x3,y3=self.__p3
        d=(y2-y3)*(x1-x3)+(x3-x2)*(y1-y3)
        a=((y2-y3)*(x-x3)+(x3-x2)*(y-y3))/d
        b=((y3-y1)*(x-x3)+(x1-x3)*(y-y3))/d
        c=1-a-b
        return a>=0 and b>=0 and c>=0


# test
r = Rectangle(5,3,0,0)
c = Circle(3,0,0)
t = Triangle(0,0,4,0,0,3)
print(r.contains(1,1), r.contains(10,10))
print(c.contains(0,0), c.contains(10,10))
print(t.contains(1,1), t.contains(3,3))
```

10. Copy the `Canvas` class from lecture to in a python file creating a `paint` module. Copy your classes from above into the module and implement paint functions. Implement a `CompoundShape` class. Create a simple drawing demonstrating that all of your classes are working.

```python
class Canvas:
    def __init__(self, w, h, bg="."):
        self.w = w
        self.h = h
        self.bg = bg
        self.clear()

    def clear(self):
        self.grid = [[self.bg for _ in range(self.w)] for _ in range(self.h)]

    def plot(self, x, y, ch="#"):
        x = int(round(x))
        y = int(round(y))
        if 0 <= x < self.w and 0 <= y < self.h:
            self.grid[self.h-1-y][x] = ch

    def show(self):
        for row in self.grid:
            print("".join(row))


class CompoundShape:
    def __init__(self):
        self.items = []

    def add(self, shape, ch="#"):
        self.items.append((shape, ch))

    def paint(self, canvas):
        for s, ch in self.items:
            for (x,y) in s.points(16):
                canvas.plot(x, y, ch)

# quick minimal shapes here just for paint demo :)

import math

class Rectangle:
    def __init__(self,L,W,x,y):
        self.L=L; self.W=W; self.x=x; self.y=y
    def points(self,n=16):
        return [(self.x,self.y),(self.x+self.L,self.y),(self.x+self.L,self.y+self.W),(self.x,self.y+self.W)]

class Circle:
    def __init__(self,r,x,y):
        self.r=r; self.x=x; self.y=y
    def points(self,n=16):
        n=max(4,min(16,int(n)))
        return [(self.x+self.r*math.cos(2*math.pi*i/n),
                 self.y+self.r*math.sin(2*math.pi*i/n)) for i in range(n)]

class Triangle:
    def __init__(self,x1,y1,x2,y2,x3,y3):
        self.p1=(x1,y1); self.p2=(x2,y2); self.p3=(x3,y3)
    def points(self,n=16):
        return [self.p1,self.p2,self.p3]


# demo
cv = Canvas(30, 15, ".")
cs = CompoundShape()
cs.add(Rectangle(8,4,2,2), "R")
cs.add(Circle(3,18,7), "O")
cs.add(Triangle(5,10, 10,10, 7,13), "T")
cs.paint(cv)
cv.show()
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self, w, h, bg="."):
        self.w = w
        self.h = h
        self.bg = bg
        self.items = []

    def add(self, shape, ch="#"):
        self.items.append((shape, ch))

    def paint(self):
        cv = Canvas(self.w, self.h, self.bg)
        for s, ch in self.items:
            for (x,y) in s.points(16):
                cv.plot(x, y, ch)
        cv.show()


# test :)
d = RasterDrawing(30, 15, ".")
d.add(Rectangle(8,4,2,2), "R")
d.add(Circle(3,18,7), "O")
d.add(Triangle(5,10, 10,10, 7,13), "T")
d.paint()

print()

d.add(Rectangle(10,1,0,0), "_")
d.paint()
```

12. Implement the ability to load/save raster drawings and demonstate that your method works. One way to implement this ability:

   * Overload `__repr__` functions of all objects to return strings of the python code that would construct the object.
   
   * In the save method of raster drawing class, store the representations into the file.
   * Write a loader function that reads the file and uses `eval` to instantiate the object.

For example:

```python
import math

class Rectangle:
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def points(self, n=16):
        x0, y0 = self.__x, self.__y
        x1, y1 = self.__x + self.__length, self.__y + self.__width
        return [(x0,y0),(x1,y0),(x1,y1),(x0,y1)]

    def __repr__(self):
        return f"Rectangle({self.__length},{self.__width},{self.__x},{self.__y})"


class Circle:
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def points(self, n=16):
        n = max(4, min(16, int(n)))
        pts = []
        for i in range(n):
            a = 2 * math.pi * i / n
            pts.append((self.__x + self.__radius*math.cos(a),
                        self.__y + self.__radius*math.sin(a))) 
        return pts

    def __repr__(self):
        return f"Circle({self.__radius},{self.__x},{self.__y})"


class Triangle:
    def __init__(self, x1, y1, x2, y2, x3, y3):
        self.__p1 = (x1, y1)
        self.__p2 = (x2, y2)
        self.__p3 = (x3, y3)

    def points(self, n=16):
        return [self.__p1, self.__p2, self.__p3]

    def __repr__(self):
        x1,y1 = self.__p1
        x2,y2 = self.__p2
        x3,y3 = self.__p3
        return f"Triangle({x1},{y1},{x2},{y2},{x3},{y3})"


class Canvas:
    def __init__(self, w, h, bg="."):
        self.w = w
        self.h = h
        self.bg = bg
        self.clear()

    def clear(self):
        self.grid = [[self.bg for _ in range(self.w)] for _ in range(self.h)]

    def plot(self, x, y, ch="#"):
        x = int(round(x))
        y = int(round(y))
        if 0 <= x < self.w and 0 <= y < self.h:
            self.grid[self.h-1-y][x] = ch

    def show(self):
        for row in self.grid:
            print("".join(row))


class RasterDrawing:
    def __init__(self, w, h, bg="."):
        self.w = w
        self.h = h
        self.bg = bg
        self.items = []

    def add(self, shape, ch="#"):
        self.items.append((shape, ch))

    def paint(self):
        cv = Canvas(self.w, self.h, self.bg)
        for shape, ch in self.items:
            for (x, y) in shape.points(16):
                cv.plot(x, y, ch)
        cv.show()

    def save(self, filename):
        with open(filename, "w") as f:
            f.write(repr((self.w, self.h, self.bg)) + "\n")
            for item in self.items:
                f.write(repr(item) + "\n")

    @staticmethod
    def load(filename):
        with open(filename, "r") as f:
            w, h, bg = eval(f.readline().strip())                            # this was just a bunch of pasting lol
            d = RasterDrawing(w, h, bg)
            for line in f:
                if line.strip():
                    shape, ch = eval(line.strip())
                    d.add(shape, ch)
        return d


# save/load
d = RasterDrawing(30, 15, ".")
d.add(Rectangle(8,4,2,2), "R")
d.add(Circle(3,18,7), "O")
d.add(Triangle(5,10, 10,10, 7,13), "T")

d.save("drawing.txt")

with open("drawing.txt", "r") as f:
    print(f.read())

d2 = RasterDrawing.load("drawing.txt")
d2.paint()
```
